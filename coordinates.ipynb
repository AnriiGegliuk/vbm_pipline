```python
import nibabel as nib
import numpy as np
from scipy.ndimage import center_of_mass
import matplotlib.pyplot as plt

# Step 1: Load the NIfTI image
image_path = 'data/TPM/01_GM.nii'
img = nib.load(image_path)

# Step 2: Get the image data as a numpy array
img_data = img.get_fdata()

# Step 3: Calculate the center of mass
com = center_of_mass(img_data)
print(f"Center of mass coordinates: {com}")

# Optional Step 4: Visualization
# Convert the center of mass coordinates to integers for visualization
com_int = tuple(map(int, com))

# Slice the image at the center of mass
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(img_data[com_int[0], :, :], cmap='gray')
axes[0].set_title('Sagittal Slice')
axes[0].plot(com_int[1], com_int[2], 'ro')  # Mark the center of mass

axes[1].imshow(img_data[:, com_int[1], :], cmap='gray')
axes[1].set_title('Coronal Slice')
axes[1].plot(com_int[0], com_int[2], 'ro')  # Mark the center of mass

axes[2].imshow(img_data[:, :, com_int[2]], cmap='gray')
axes[2].set_title('Axial Slice')
axes[2].plot(com_int[0], com_int[1], 'ro')  # Mark the center of mass

plt.show()
```
